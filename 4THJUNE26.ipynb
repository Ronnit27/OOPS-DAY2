```python
# function of prime numbers
print("the prime numbers are between 1 to 10: ")
for i in range (1,11):
  if (i == 1):
    continue
  for j in range (2,i):
    if (i%j == 0):
      break
  else:
    print(i)
```

```python
print("the prime numbers are between 10 to 20: ")
for num in range(10,20):
  print(type(num))
  # Primes are greater than 1
  if num > 1:
    # Check for factors from 2 up to num-1
    for i in range(2, num):
      if (num % i) == 0:
        break # If a factor is found, it's not prime
    else:
      print(num) # If no factors were found, it is a prime number
```

```python
def average(*numbers_tuple):
  # numbers_tuple will be a tuple of integers
  sum_of_numbers = 0
  for num in numbers_tuple:
    sum_of_numbers = sum_of_numbers + num
  print("average is", sum_of_numbers / len(numbers_tuple))

# Get input from the user and convert to integers
num1 = int(input("enter the first number: "))
num2 = int(input("enter the second number: "))
num3 = int(input("enter the third number: "))

# Pass all three integer numbers to the average function
average(num1, num2, num3)
```

```python
# RECURSION
# factorial
# FACTORIAL OF 5 = (5*4*3*2*1) FORMULA
# factorial(n)= n * factocial (n-1)
def factorial(n):
  if (n==0 or n==1):
    return 1
  else:
    return n * factorial(n-1)
print(factorial(5))
```

```python
print(factorial(10))
print (factorial(3))
```

```python
def fibonacci(n):
  if n <= 0:
    return 0
  elif n == 1:
    return 1
  else:
    return fibonacci(n-1) + fibonacci(n-2)
print(6,4,5)
```
